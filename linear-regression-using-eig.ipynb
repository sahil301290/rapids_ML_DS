```python
import cudf as pd
from sklearn.datasets import load_boston
boston = load_boston()
df = pd.DataFrame(data=boston.data, columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
        'TAX', 'PTRATIO', 'B', 'LSTAT'])
df['target'] = boston.target
df.head(5)
```

```python
import cupy as np
X=df.iloc[:,0:13]
Y=df.iloc[:,13]
Y=np.array(Y)
Y=Y.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X.as_matrix())
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from cuml import LinearRegression
```

```python
lr = LinearRegression(fit_intercept = True, normalize = False,
                      algorithm = "eig")
```

```python
reg = lr.fit(X_train,y_train)
```

```python
print("Coefficients:")
print(reg.coef_)
print("Intercept:")
print(reg.intercept_)
```

```python
preds = lr.predict(X_test)
```

```python
print("Predictions:")
print(preds)
```

```python
import cuml
cuml.metrics.regression.mean_squared_error(y_test,preds)
```

```python
cuml.metrics.regression.r2_score(y_test,preds)
```

```python
cuml.metrics.regression.mean_absolute_error(y_test,preds)
```
